# Question 1: How many finance lending and blockchain clients does the organization have?

In this step, we will determine the number of clients in two specific industries: "Finance Lending" and "Blockchain." We will count the number of distinct clients in each of these industries and visualize the results in a bar plot.


```python
# Filter for finance lending and blockchain clients
finance_lending_clients = industry_client_details[industry_client_details['industry'] == 'Finance Lending']
blockchain_clients = industry_client_details[industry_client_details['industry'] == 'Block Chain']

# Count the distinct clients
finance_lending_count = finance_lending_clients['client_id'].nunique()
blockchain_count = blockchain_clients['client_id'].nunique()

print(f"Number of Finance Lending clients: {finance_lending_count}")
print(f"Number of Blockchain clients: {blockchain_count}")
```

```python
# Count the number of clients in Finance Lending and Blockchain industries
industry_counts = industry_client_details['industry'].value_counts()

# Plotting the count of clients by industry
plt.figure(figsize=(8, 5))
industry_counts.plot(kind='bar', color=['blue', 'orange'])
plt.title('Number of Clients by Industry')
plt.xlabel('Industry')
plt.ylabel('Number of Clients')
plt.xticks(rotation=45)
plt.show()
```

# Question 2: Which industry in the organization has the highest renewal rate?

In this step, we calculate the renewal rate for each industry by dividing the number of renewed subscriptions by the total number of subscriptions in that industry. We will visualize the renewal rates for each industry using a bar plot.


```python
# Merge subscription and industry client details to get the industry for each subscription
subscription_with_industry = pd.merge(subscription_information, industry_client_details, on='client_id', how='left')

# Calculate renewal rate for each industry
industry_renewals = subscription_with_industry.groupby('industry').agg(
    renewal_rate=('renewed', 'mean')
).sort_values(by='renewal_rate', ascending=False)

print(f"Industry with the highest renewal rate: {industry_renewals.head(1)}")
```

```python
# Calculate renewal rates for each industry
industry_renewals = subscription_with_industry.groupby('industry').agg(
    renewal_rate=('renewed', 'mean')
).sort_values(by='renewal_rate', ascending=False)

# Plotting renewal rates by industry
plt.figure(figsize=(10, 6))
sns.barplot(x=industry_renewals.index, y=industry_renewals['renewal_rate'], palette='viridis')
plt.title('Renewal Rate by Industry')
plt.xlabel('Industry')
plt.ylabel('Renewal Rate')
plt.xticks(rotation=45)
plt.show()
```

# Question 3: What was the average inflation rate when their subscriptions were renewed?

In this step, we will calculate the average inflation rate for the periods when subscriptions were renewed. We will filter the dataset to include only the renewed subscriptions and then calculate the average inflation rate. Finally, we will visualize the inflation rate distribution using a box plot.


```python
# Merge subscription information with financial information to get inflation rates for renewal periods
subscription_with_inflation = pd.merge(subscription_with_industry, financial_information, left_on='start_date', right_on='start_date', how='left')

# Filter for renewed subscriptions and calculate average inflation rate
average_inflation_rate = subscription_with_inflation[subscription_with_inflation['renewed'] == True]['inflation_rate'].mean()

print(f"Average inflation rate during renewal: {average_inflation_rate}")
```

```python
# Filter for renewed subscriptions
renewed_subscriptions = subscription_with_inflation[subscription_with_inflation['renewed'] == True]

# Plotting the boxplot for inflation rates during renewals
plt.figure(figsize=(8, 5))
sns.boxplot(data=renewed_subscriptions, x='industry', y='inflation_rate', palette='Set2')
plt.title('Inflation Rate During Subscription Renewals')
plt.xlabel('Industry')
plt.ylabel('Inflation Rate')
plt.xticks(rotation=45)
plt.show()
```

# Question 4: What is the median amount paid each year for all payment methods?

In this step, we will calculate the median amount paid for each payment method, grouped by the year of payment. We will visualize the median payment amounts by year for each payment method using a bar plot.


```python
# Convert payment date to datetime format
payment_information['payment_date'] = pd.to_datetime(payment_information['payment_date'], format='%m/%d/%Y')

# Extract the year from the payment date
payment_information['payment_year'] = payment_information['payment_date'].dt.year

# Calculate median payment amount per year for each payment method
median_payment_per_method = payment_information.groupby(['payment_method', 'payment_year'])['amount_paid'].median().reset_index()

# Display the results
print(median_payment_per_method)
```

```python
# Calculate the median payment per year for each payment method
median_payment_per_method = payment_information.groupby(['payment_method', 'payment_year'])['amount_paid'].median().reset_index()

# Plotting the median payment amount by year and payment method
plt.figure(figsize=(10, 6))
sns.barplot(data=median_payment_per_method, x='payment_year', y='amount_paid', hue='payment_method')
plt.title('Median Payment Amount by Payment Method and Year')
plt.xlabel('Year')
plt.ylabel('Median Payment Amount')
plt.xticks(rotation=45)
plt.show()
```
